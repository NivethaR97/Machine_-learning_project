##### 2) Are there any missing(NaN) and invalid values in the dataset, and how are they handled?

##### a)Removing invalid entries

```python
df["thal"].value_counts()
```

```python
df[df["thal"]==3]
```

```python
#invalid values in thal=0
df=df[df["thal"]!=3]
```

```python
df["ca"].value_counts()
```

```python
#invalid values in ca
df=df[df["ca"]!=4]
```

##### 3) Are there any duplicate rows in the dataset, and how are they handled?

```python
#Checking for redundant rows
wrong_df=df["age"]=="age"
df[wrong_df]#no redundant rows
```

```python
#Check for duplicate rows
df[df.duplicated()]
```

```python
df = df.drop_duplicates() #drop the duplicates
```
